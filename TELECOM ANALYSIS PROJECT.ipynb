```python
df['Total_Mins'] = df['Day Mins'] + df['Eve Mins'] + df['Night Mins'] + df['Intl Mins']
```

```python
df['Total_Calls'] = df['Day Calls'] + df['Eve Calls'] + df['Night Calls'] + df['Intl Calls']
```

```python
df ['Total_charge']= df['Day Charge'] + df['Eve Charge'] + df['Night Charge'] + df['Intl Charge']
```

```python
df['Avg_Call_Duration'] = np.where(df['Total_Calls']==0, 0, df['Total_Mins'] / df['Total_Calls'])
```

```python
df.head()
```

```python
df.to_csv(r'C:\Users\itsra\Downloads\'Cleaned_Telecom_Call_Data.csv', index=False)
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Total_charge'], bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Total Charges')
plt.xlabel('Total Charges $')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Total_Calls'], df['Total_charge'], alpha=0.5, color='orange')
plt.title('Total Calls vs Total Charges')
plt.xlabel('Total Calls Made')
plt.ylabel('Total Charges ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Total_Mins'], df['Total_charge'], alpha=0.5, color='green')
plt.title('Total Minutes Used vs Total Charges')
plt.xlabel('Total Minutes Used')
plt.ylabel('Total Charges ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

calls_count = df['CustServ Calls'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
plt.bar(calls_count.index, calls_count.values, color='purple', edgecolor='black')
plt.title('Customer Service Calls Distribution')
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Churn', y='Day Mins', data=df, estimator='mean', color='salmon')

plt.title('Average Daytime Minutes vs Churn Status')
plt.xlabel('Churn Status (True = Left)')
plt.ylabel('Average Daytime Minutes')
plt.show()
```
